# Plot QQ-norm graph from scratch

```python
import numpy as np
import seaborn as sns
import pylab
import statsmodels.api as sm
from scipy.stats import norm
```

```python
# generate 100 points of test data = N(100, 35) + eps(0, 1)
test_data = np.random.normal(loc=100, scale=35, size=100) + np.random.normal(loc=0, scale=0, size=100)
```

```python
# Plot QQ-norm graph
def qq_norm(data):
  perc = [i/(len(data)+1) for i in range(1, len(data)+1)] # calculate percentile of every sample
  norm_inv = norm.ppf(perc) # calculate the inverse of the normal distribution
  sns.scatterplot(x=norm_inv, y=sorted(data))
```

```python
qq_norm(test_data)
```

```python
# draw the same plot from 'statsmodels' library
sm.qqplot(test_data, line = 'q')
pylab.show()
```

**Conclusion:** Our plot coincides with the statsmodels's plot
